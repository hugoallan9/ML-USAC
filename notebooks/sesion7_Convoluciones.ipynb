# Ejemplos de convoluciones en imágenes
En esta libreta se trabajan algunos ejemplos de convoluciones en imágenes.
La biblioteca python que se utiliza es OpenCV.

```python
#importación de las bibliotecas
import cv2
import numpy as np
import matplotlib.pyplot as plt
import requests
```

```python
#Bajar la imagen de internet
url = "https://cdn.pixabay.com/photo/2015/01/07/16/37/wood-591631_960_720.jpg"
r = requests.get(url)
with open('test.jpg', 'wb') as f:
    f.write(r.content)
```

```python
#Guardar la imagen en una variable
image = cv2.imread('test.jpg')
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.axis("off")
plt.imshow(image)
```

## Primer filtro
Con la ayuda de la biblioteca numpy haremos un primer filtro (o kernel).

```python
kernel = np.array([[0,0,0],[0,1,0], [0,0,0]])
```

```python
#Aplicando la convolución
filtro = cv2.filter2D(src=image, ddepth=-1, kernel=kernel)
```

```python
plt.figure(figsize=(18,19))
plt.subplot(121)
plt.imshow(image)
plt.title("Imagen original")
plt.axis("off")
plt.subplot(122)
plt.imshow(filtro)
plt.title("Imagen con filtro")
plt.axis("off")
```

## Filtro para hacer borrosa una imagen

```python
kernel = np.ones((3,3), np.float32) / 9
```

```python
filtro = cv2.filter2D(src=image, ddepth=-1, kernel=kernel)
```

```python
plt.figure(figsize=(18,19))
plt.subplot(121)
plt.imshow(image)
plt.title("Imagen original")
plt.axis("off")
plt.subplot(122)
plt.imshow(filtro)
plt.title("Imagen con filtro borroso")
plt.axis("off")
```

```python
img_blur = cv2.blur(src=image, ksize=(5,5))
```

```python
plt.figure(figsize=(18,19))
plt.subplot(121)
plt.imshow(image)
plt.title("Imagen original")
plt.axis("off")
plt.subplot(122)
plt.imshow(img_blur)
plt.title("Imagen con filtro borroso")
plt.axis("off")
```
